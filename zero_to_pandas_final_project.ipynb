## Exploratory Analysis and Visualization

**TODO** - In this part i am going to Explore the data and also conclude some information using graphs.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for seaborn
sns.set(style="whitegrid")
```

```python
# Explore the distribution of salaries

plt.figure(figsize=(12, 6))
sns.histplot(analysis_df['salary_in_usd'], bins=30, kde=True)
plt.title('Distribution of Salaries')
plt.xlabel('Salary in USD')
plt.ylabel('Frequency')
plt.show()
```

```python
# Explore the distribution of job categories

plt.figure(figsize=(12, 6))
sns.countplot(y='job_category', data=analysis_df, order=analysis_df['job_category'].value_counts().index)
plt.title('Distribution of Job Categories')
plt.xlabel('Number of Employees')
plt.ylabel('Job Category')
plt.show()
```

It seems like that Data Science and Research employe are the most

```python
# Explore the average salary by experience level

plt.figure(figsize=(12, 6))
sns.barplot(x='experience_level', y='salary_in_usd', data=analysis_df, order=['Entry-level', 'Mid-level', 'Senior'])
plt.title('Average Salary by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Average Salary in USD')
plt.show()
```

As we can see above the Average salary of Senior Level is above 140000 USD

```python
# Explore the distribution of employment types

plt.figure(figsize=(12, 6))
sns.countplot(y='employment_type', data=analysis_df, order=analysis_df['employment_type'].value_counts().index)
plt.title('Distribution of Employment Types')
plt.xlabel('Number of Employees')
plt.ylabel('Employment Type')
plt.show()
```

We can see the employment type is full-time only

```python
# Explore the distribution of company sizes

plt.figure(figsize=(12, 6))
sns.countplot(y='company_size', data=analysis_df, order=analysis_df['company_size'].value_counts().index)
plt.title('Distribution of Company Sizes')
plt.xlabel('Number of Employees')
plt.ylabel('Company Size')
plt.show()
```

We can see the majority of the company size is  medium

```python
#Line Plot - Salary Trends Over Years:

plt.figure(figsize=(12, 6))
sns.lineplot(x='work_year', y='salary_in_usd', data=analysis_df, estimator='mean', err_style=None)
plt.title('Average Salary Trends Over Years')
plt.xlabel('Work Year')
plt.ylabel('Average Salary in USD')
plt.show()

```

```python

#Box Plot - Salary Distribution by Job Category:

plt.figure(figsize=(12, 6))
sns.boxplot(x='job_category', y='salary_in_usd', data=analysis_df)
plt.title('Salary Distribution by Job Category')
plt.xlabel('Job Category')
plt.ylabel('Salary in USD')
plt.xticks(rotation=45)
plt.show()

```

```python
#Count Plot - Employee Residence Distribution:
plt.figure(figsize=(12, 6))
sns.countplot(x='employee_residence', data=analysis_df, order=analysis_df['employee_residence'].value_counts().index)
plt.title('Distribution of Employee Residence')
plt.xlabel('Employee Residence')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45)
plt.show()
```

## Asking and Answering Questions


### Q1. What is the average salary for each job category??

```python

# Calculate the average salary by job category

average_salary_by_category = analysis_df.groupby('job_category')['salary_in_usd'].mean().sort_values()

# Plotting
plt.figure(figsize=(12, 6))
sns.barplot(x=average_salary_by_category.index, y=average_salary_by_category.values)
plt.title('Average Salary by Job Category')
plt.xlabel('Job Category')
plt.ylabel('Average Salary in USD')
plt.xticks(rotation=45, ha='right')
plt.show()
```

### Q2. What is the distribution of experience levels in the dataset?

```python


# Calculate the distribution of experience levels
experience_level_distribution =analysis_df['experience_level'].value_counts()

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x=experience_level_distribution.index, y=experience_level_distribution.values)
plt.title('Distribution of Experience Levels')
plt.xlabel('Experience Level')
plt.ylabel('Number of Employees')
plt.show()
```

### Q3. What is the average salary for each experience level?

```python


# Calculate the average salary by experience level
average_salary_by_experience = analysis_df.groupby('experience_level')['salary_in_usd'].mean().sort_values()

# Plotting
plt.figure(figsize=(12, 6))
sns.barplot(x=average_salary_by_experience.index, y=average_salary_by_experience.values)
plt.title('Average Salary by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Average Salary in USD')
plt.xticks(rotation=45, ha='right')
plt.show()
```

## Q4.  How many employees work in each employment type?

```python

# Calculate the number of employees per employment type
employees_per_employment_type = analysis_df['employment_type'].value_counts()

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x=employees_per_employment_type.index, y=employees_per_employment_type.values)
plt.title('Number of Employees per Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Number of Employees')
plt.show()
```

Employement type is 100% is full-time

## Q5.  What is the average salary for employees working in-person vs. remote?

```python

# Calculate the average salary by work setting
average_salary_by_work_setting = analysis_df.groupby('work_setting')['salary_in_usd'].mean().sort_values()

# Plotting
plt.figure(figsize=(12, 6))
sns.barplot(x=average_salary_by_work_setting.index, y=average_salary_by_work_setting.values)
plt.title('Average Salary by Work Setting')
plt.xlabel('Work Setting')
plt.ylabel('Average Salary in USD')
plt.xticks(rotation=45, ha='right')
plt.show()
```

## Q6.  How does the average salary vary for different job categories over the years?

```python


average_salary_by_category_over_years = analysis_df.groupby(['work_year', 'job_category'])['salary_in_usd'].mean().unstack()
plt.figure(figsize=(12, 6))
average_salary_by_category_over_years.plot(marker='o')
plt.title('Average Salary by Job Category Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Salary in USD')
plt.legend(title='Job Category', bbox_to_anchor=(1, 1))
plt.show()
```
